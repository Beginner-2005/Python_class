```python
# function without argument and no return type

def cal():
    l=int(input("Enter length:"))    #local variable (all variable inside funcn)
    b=int(input("Enter breadth:"))    #local variable 
    h=int(input("Enter height:"))    #local variable 
    a=l*b                           #local variable 
    v=a*h                           #local variable 
    print(a)
    print(v)
    
cal()
```

```python
# function with argument and no return type

def cal(l,b,h):         #<-- parameter
    a=l*b                           #local variable 
    v=a*h                           #local variable 
    print(a)
    print(v)

l=int(input("Enter length:"))    #global variable 
b=int(input("Enter breadth:"))    #global variable 
h=int(input("Enter height:"))    #global variable 

cal(l,b,h)               #l,b,h argument
```

```python
# function with argument and no return type
# parameter and argument must have same no of variable to pass input


def cal(x,y,z):         #<-- parameter
    a=x*y                           #local variable 
    v=a*z                           #local variable 
    print(a)
    print(v)

l=int(input("Enter length:"))    #global variable 
b=int(input("Enter breadth:"))    #global variable 
h=int(input("Enter height:"))    #global variable 

cal(l,b,h)               #l,b,h argument
```

```python
def area(l,b):
    a=l*b
    print("Area:",a)
    
def volume(l,b,h):
    v=l*b*h
    print("Volume:",v)
    
l=int(input("Enter length:"))    #global variable 
b=int(input("Enter breadth:"))    #global variable 
h=int(input("Enter height:"))    #global variable 

area(l,b)                   #l,b argument
volume(l,b,h)               #l,b,h argument
```

```python
# function without argument and return type
def cal():
    l=int(input("Enter l:"))
    b=int(input("Enter b:"))
    h=int(input("Enter h:"))
    a=l*b
    v=a*h
    return a,v
x=cal()
print(x)
area,volume=x
print("Area=",area)
print("Volume=",volume)
```

```python
# function with argument and return type
def cal(l,b,h):
    a=l*b
    v=a*h
    return a,v

l=int(input("Enter l:"))
b=int(input("Enter b:"))
h=int(input("Enter h:"))

x=cal(l,b,h)
area,volume=x
print("Area=",area)
print("Volume=",volume)
```

```python
# filter, map
```

```python
a=[1,2,3,4,5,6,7,8,9,0]
list(filter(lambda x:x%2==0,a))
```

```python
a=["apple","Ball","CAT","Dog","FISH"]
list(filter(lambda x:x.isupper(),a))
```

```python
a=["apple","Ball","CAT","Dog","FISH"]
list(filter(lambda x:x[0].isupper(),a))
```

```python
#map function    (using arguement to perform task)
a=["Apple","Ball","CAT","Dog","FISH"]
list(map(lambda x:x.lower(),a))
```

```python
a=[1,2,3,4,5,6,7,8,9]
print(list(map(lambda x:x**2,a)))
```

```python
# Calculator
def calc():
    a=int(input("Enter a:"))
    b=int(input("Enter b:"))
    o=input("Enter an operator(+,-,*,/)")
    if o=="+":
        c=a+b
    elif o=="-":
        c=a-b
    elif o=="*":
        c=a*b
    elif o=="/":
        c=a/b
    else:
        print("The operator is wrong")
    return c

print(calc())
```
